```python
!pip install pandas matplotlib seaborn --quiet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io


print("üìÇ Selecciona el archivo 'train.csv' desde tu equipo:")
uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['train.csv']))

print("‚úÖ Archivo cargado correctamente. Dimensiones del dataset:", df.shape)

df.head()
```

```python
plt.style.use('ggplot')
sns.set_palette('Set2')


print("\nüîπ Informaci√≥n del dataset:")
df.info()

print("\nüîπ Valores faltantes por columna:")
display(df.isnull().sum())


if 'RENDIMIENTO_GLOBAL' in df.columns:
    print("\nüìä Distribuci√≥n de la variable objetivo (RENDIMIENTO_GLOBAL):")
    display(df['RENDIMIENTO_GLOBAL'].value_counts())


    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x='RENDIMIENTO_GLOBAL', order=df['RENDIMIENTO_GLOBAL'].value_counts().index)
    plt.title("Distribuci√≥n de clases - RENDIMIENTO_GLOBAL")
    plt.xlabel("Categor√≠a de rendimiento")
    plt.ylabel("N√∫mero de estudiantes")
    plt.show()
else:
    print("‚ö†Ô∏è No se encontr√≥ la columna 'RENDIMIENTO_GLOBAL' en el dataset.")


possible_cols = [col for col in df.columns if 'local' in col.lower()]
if possible_cols:
    col_localidad = possible_cols[0]
    print(f"\nüìä Distribuci√≥n por {col_localidad}:")
    display(df[col_localidad].value_counts().head(10))

    plt.figure(figsize=(8,4))
    df[col_localidad].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top 10 localidades con m√°s estudiantes ({col_localidad})")
    plt.xlabel(col_localidad)
    plt.ylabel("Cantidad de estudiantes")
    plt.show()
else:
    print("\n‚ö†Ô∏è No se encontr√≥ una columna relacionada con 'localidad' en el dataset.")


print("\nüìà Estad√≠sticas descriptivas de variables num√©ricas:")
display(df.describe())
```
